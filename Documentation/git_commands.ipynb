# Git Commands

## Initialize
### Setup: 
[Adding an existing project to GitHub using the command line](https://docs.github.com/en/github/importing-your-projects-to-github/adding-an-existing-project-to-github-using-the-command-line)

- **Initialize Git by creating a .git file**  
`git init`

- **Set the new remote**  
`git remote add origin <REMOTE_URL>`

- **Verify the new remote URL**  
`git remote -v`

- **Set your username**  
`git config --global user.name "FIRST_NAME LAST_NAME"`

- **Set your email address**  
`git config --global user.email "xxxxxx1@gmail.com"`

## Modify/Add/Change/Commit/Push/etc...
- **Add all files in the current directory**  
`git add .`

- **Commit changes with a message**  
`git commit -m 'project change'`

- **Check the status of files**  
`git status`

- **See commit history**  
`git log`

- **Add a file**  
`touch style.css`

- **Remove a file from tracking (ignore it)**  
`git rm --cached style.css`

- **Alternatively, create a .gitignore file and include all the files you want to ignore**  

- **Revert files to last commit**  
`git checkout -- .`

- **Create a new branch**  
`git branch branchname`

- **List all branches**  
`git branch --list`

- **Switch to a branch**  
`git checkout branchname`

- **Merge branches**  
`git merge branchname`

- **List remote connections**  
`git remote`

- **Push changes to the master branch**  
`git push origin master`
  
`git push -u origin master`

- **Clone a repository from GitHub**  
`git clone <link>`

- **Install all packages from package.json (if using npm)**  
`npm install`

- **Remove git from a directory**  
`rm -rf .git`

## Working With a Team
- **Add a remote URL for each team member**  
`git remote add <url>`
